```python
from collections import Counter

def wordCount(text):
    dictionary = {};
    list_of_words = re.split("\.+", re.sub('[^a-zA-Z0-9]', '.', text).strip('.'))
    wordsCounter = Counter(list_of_words)
    for word in list_of_words:
        dictionary [word]= wordsCounter[word]

    return dictionary
    ## end ##
```
